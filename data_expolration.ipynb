```python

import os
import pandas as pd
import numpy as np

from typing import Dict
from os.path import isfile, join
```

```python
DATASET_DIR = "dataset"

def load_dataset(dataset_dir: os.PathLike) -> Dict[str, pd.DataFrame]:
    # Gets paths of all the file in a directory
    def list_dataset_dir(dataset_dir: os.PathLike):
        return [join(dataset_dir, file) for file in os.listdir(dataset_dir)]

    def is_csv(file: os.PathLike):
        return isfile(file) and file.endswith(".csv")

    # Given a filename in the form `filename.ext`, returns `filename`.
    def file_name(filename: os.PathLike):
        # The placeholder is the extension of the file.
        filename, _ = filename.split(os.path.sep)[-1].split('.')
        return filename

    # Fetches the list of CSV files in the given directory, Then loads the CSVs
    # into dataframes. Then aggregates all the dataframes into one big
    # dictionary with the filename (without the extension) as the key and the
    # data as the value.
    csv_filenames = [file for file in list_dataset_dir(dataset_dir) if is_csv(file)]

    data = [pd.read_csv(file) for file in csv_filenames]
    col_names = [file_name(file) for file in csv_filenames]

    return { col_name: data for col_name, data in zip(col_names, data)}
```

```python
dataset = load_dataset(DATASET_DIR)
```
